```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
```

```python
# Load the dataset
df = pd.read_csv('data.csv')
```

```python
# 1. Bar Graph: Average MSRP by Vehicle Size
plt.figure(figsize=(10, 6))
sns.barplot(x='Vehicle Size', y='MSRP', data=df, estimator='mean')
plt.title('Average MSRP by Vehicle Size')
plt.show()
```

```python
# 2. Histogram: Distribution of Engine HP
plt.figure(figsize=(10, 6))
plt.hist(df['Engine HP'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Engine HP')
plt.xlabel('Engine HP')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Pie Chart: Distribution of Vehicle Sizes
vehicle_size_counts = df['Vehicle Size'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(vehicle_size_counts, labels=vehicle_size_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Vehicle Sizes')
plt.show()
```

```python
# 4. Line Graph: highway MPG over the Year for a specific Driven_Wheels 
df_awd = df[df['Driven_Wheels'] == 'all wheel drive']
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='highway MPG', data=df_awd, marker='o')
plt.title('Highway MPG over the Years for All Wheel Drive')
plt.xlabel('Year')
plt.ylabel('Highway MPG')
plt.show()
```

```python
# 5. Scatter Plot: Engine Cylinders vs. Popularity
plt.figure(figsize=(10, 6))
plt.scatter(df['Engine Cylinders'], df['Popularity'], alpha=0.6, color='purple')
plt.title('Engine Cylinders vs. Popularity')
plt.xlabel('Engine Cylinders')
plt.ylabel('Popularity')
plt.show()
```

```python
# 6. Bubble Chart: Engine HP vs. MSRP with bubble size based on Popularity
plt.figure(figsize=(10, 6))
plt.scatter(df['Engine HP'], df['MSRP'], s=df['Popularity'] / 50, alpha=0.5)
plt.title('Engine HP vs. MSRP with Popularity Bubble Size')
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.show()
```

```python
# 7. Violin Plot: Distribution of MSRP by Vehicle Size
plt.figure(figsize=(10, 6))
sns.violinplot(x='Vehicle Size', y='MSRP', data=df)
plt.title('Distribution of MSRP by Vehicle Size')
plt.show()
```

```python
# 8. Box Plot: Distribution of Engine HP by Vehicle Size
plt.figure(figsize=(10, 6))
sns.boxplot(x='Vehicle Size', y='Engine HP', data=df)
plt.title('Distribution of Engine HP by Vehicle Size')
plt.show()
```

```python
# 9. Area Chart: Cumulative count of vehicles over the years
df['Year'] = df['Year'].astype(str)
year_counts = df['Year'].value_counts().sort_index()
plt.figure(figsize=(15, 6))
plt.fill_between(year_counts.index, year_counts.values, color="skyblue", alpha=0.4)
plt.plot(year_counts.index, year_counts.values, color="Slateblue", alpha=0.6, linewidth=2)
plt.title('Cumulative Count of Vehicles Over the Years')
plt.xlabel('Year')
plt.ylabel('Count of Vehicles')
plt.show()
```

```python
# 10. Tree Map: Count of Vehicle Style by Make
make_vehicle_style = df.groupby('Make')['Vehicle Style'].count().reset_index()
plt.figure(figsize=(10, 6))
squarify.plot(sizes=make_vehicle_style['Vehicle Style'], label=make_vehicle_style['Make'], alpha=.8, color=sns.color_palette('Set3'))
plt.axis('off')
plt.title('Count of Vehicle Style by Make')
plt.show()
```
